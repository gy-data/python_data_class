```python
import matplotlib.pyplot as plt
import numpy as np
```

## 기본 그래프(Plot)

```python
plt.plot([1,2,3,4,5,6,5,4,3,2,1])
plt.show()
```

```python
# sin 함수

x = np.arange(0,12,0.01) #0~12까지 0.01 간격
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

## Title 설정

```python
plt.plot(x,y)
plt.title("sin graph")
plt.show()
```

## grid 설정

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

## x, y 라벨

```python
plt.plot(x,y)
plt.xlabel('time')
plt.ylabel('Amplotude') #진폭
plt.show()
```

## 범례(legend)

```python
plt.plot(x,np.sin(x), label="sin")
plt.plot(y,np.sin(y), label="cos")
plt.grid()
plt.legend(loc=(1.05, 0.05)) # loc
plt.show()
```
