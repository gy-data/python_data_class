## 구글 뉴스 크롤링(10p까지)

- 목적 : 구글 누스를 10페이지까지 긁어오는 것

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
path ="/Users/young/Desktop/chromedriver"
browser = webdriver.Chrome(path)
```

```python
import time

search_word=input("크롤링하고자 하는 단어를 입력하세요.")
page = int(input("크롤링하고자 하는 페이지수를 입력하세요."))

url = "https://www.google.com/search?q={}&tbm=nws&ei=YKxYY8yEFoyMoASJ9qlo&start={}&sa=N&ved=2ahUKEwjMtIXi_Pz6AhUMBogKHQl7Cg0Q8tMDegQIARAG&biw=1200&bih=707&dpr=2"

for i in range(0, page*10, 10):
    fin_url = url.format(search_word, i)
    browser.get(fin_url)
    
    # 사람인 척 하기 위한 장치들
    time.sleep(3) # 3초 뒤에 다음 페이지로 넘어감.
    browser.implicitly_wait(10) # 10초 기다려봐. 안 되면 오류를 뱉어내고, 만약에 그 전에 로딩되면 진행.
    
    titles = browser.find_elements(By.CLASS_NAME,'mCBkyc')
    for j in titles:
        print(j.text)
```
